```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('https://raw.githubusercontent.com/tewei0328/data/master/TX11_day.csv')
```

```
df.head()
```

```
df=df[::-1]
```

```
df['ma5']=df['close'].rolling(5).mean()
df['ma20']=df['close'].rolling(20).mean()
```

```
df.index=df['date']
```

```
df.head(30)
```

```
#df[['close','ma5','ma20']].plot(figsize=(20,5), rot=30, x=df.date)
df[['close','ma5','ma20']].plot(figsize=(20,5), rot=30)
```

```
x='ma5'
#y='ma20'
#x='close'
#x='open'
#y='ma5'
y='ma20'
```

```
#
for i in range(len(df)):
    if i>1:
        if df.ix[i-1,x]>df.ix[i-1,y] and df.ix[i,x]<=df.ix[i,y]:
            df.ix[i,'sk']=-1
            print(i)
```

```
for i in range(len(df)):
    if i>1:
        if df.ix[i-1,x]<df.ix[i-1,y] and df.ix[i,x]>=df.ix[i,y]:
            df.ix[i,'sk']=1
            print(i)
```

```
df.head(50)
```

```
plt.figure(figsize=(150,10))
plt.xticks(rotation=90)#
plt.plot(df[['close','ma5','ma20']])
plt.show()
```

```
df['position']=''
for i in range(len(df)):
    if i>0:
        if df.ix[i,'sk']==1:
           df.ix[i,'position']=1
        if df.ix[i,'sk']==-1:
           df.ix[i,'position']=-1  
        if df.ix[i,'sk']!=1 and df.ix[i,'sk']!=-1:
           df.ix[i,'position']=df.ix[i-1,'position']
```

```
for i in range(len(df)):
    if i>0:
        if df.ix[i-1,'position']==1:
            df.ix[i,'profit']=df.ix[i,'close']-df.ix[i-1,'close']
        if df.ix[i-1,'position']==-1:
            df.ix[i,'profit']=-df.ix[i,'close']+df.ix[i-1,'close'] 
```

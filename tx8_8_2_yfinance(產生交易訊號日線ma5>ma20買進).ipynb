```
#
for i in range(len(df)):
    if i>1:
        if df.loc[i-1,x]>df.loc[i-1,y] and df.loc[i,x]<=df.loc[i,y]:
            df.loc[i,'sk']=-1
            print(i)
```

```
for i in range(len(df)):
    if i>1:
        if df.loc[i-1,x]<df.loc[i-1,y] and df.loc[i,x]>=df.loc[i,y]:
            df.loc[i,'sk']=1
            print(i)
```

```
plt.figure(figsize=(100,20))
plt.xticks(rotation=90)#设置时间标签显示格式
plt.plot(df[['close','ma5','ma20']])
plt.show()
```

```
df['position']=''
for i in range(len(df)):
    if i>0:
        if df.loc[i,'sk']==1:
           df.loc[i,'position']=1
        if df.loc[i,'sk']==-1:
           df.loc[i,'position']=-1  
        if df.loc[i,'sk']!=1 and df.loc[i,'sk']!=-1:
           df.loc[i,'position']=df.loc[i-1,'position']
```

```
for i in range(len(df)):
    if i>0:
        if df.loc[i-1,'position']==1:
            df.loc[i,'profit']=df.loc[i,'close']-df.loc[i-1,'close']
        if df.loc[i-1,'position']==-1:
            df.loc[i,'profit']=-df.loc[i,'close']+df.loc[i-1,'close'] 
```

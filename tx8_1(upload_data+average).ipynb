```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('TX00_day.csv')
```

```
df.head()
```

```
df['ma5']=df['close'].rolling(5).mean()
df['ma20']=df['close'].rolling(20).mean()
```

```
df.head(20)
```

```
for i in range(len(df)):
    if i>1:
        if df.ix[i-1,'ma5']>df.ix[i-1,'ma20'] and df.ix[i,'ma5']<=df.ix[i,'ma20']:
            df.ix[i,'sk']=-1
        if df.ix[i-1,'ma5']<df.ix[i-1,'ma20'] and df.ix[i,'ma5']>=df.ix[i,'ma20']:
            df.ix[i,'sk']=1
```

```
df['close'].plot()
```

```
df[['close','ma5','ma20']].plot()
```

```
df[['close','ma5','ma20']].plot(figsize=(20,5))
```
